# 2020_セジロウンカ日別誘殺数

## tabulaによる表の抽出

```python
dfs = tabula.read_pdf(
    "香川県農業試験場病害虫防除所_業務年報_2020.pdf",
    pages = '79',
    lattice=True
    )

# for df in dfs[0:-2]: # list内の1要素が1つの表に対応（dfs[0]と同義）
#     display(df)
# print(type(dfs[0]))
df = dfs[0]

df.drop([0], inplace=True)

df.columns = [
    '日',
    '4月_メス', '4月_オス',
    '5月_メス', '5月_オス',
    '6月_メス', '6月_オス',
    '7月_メス', '7月_オス',
    '8月_メス', '8月_オス',
    '9月_メス', '9月_オス',
    '10月_メス', '10月_オス',
    '11月_メス', '11月_オス'
    ]

df
```

```python
# すべての列を対象に分割
for col in df.columns:
    df[col] = df[col].str.split('\r')

# 各行で「月日」の長さに合わせて他の列をパディング
for col in df.columns:
    max_len = df['日'].map(len) # 各行の「日」の長さ
    new_column = []
    for lst, ml in zip(df[col], max_len):
        if len(lst) < ml:
            new_lst = lst + [np.nan] * (ml - len(lst))
        else:
            new_lst = lst
        new_column.append(new_lst)
    df[col] = new_column

# 各セルがリストになったので，まとめて縦方向にexplode
df = df.explode(list(df.columns), ignore_index=True)

df
```

## 誤記チェック

```python
# print(df.dtypes)
# print('---')
df = df.astype('float64') # object -> float64
# print(df.dtypes)

df.sum()
```

## CSVとして保存

```python
df.to_csv("セジロウンカ日別誘殺数_2020.csv")
```

# 2024_セジロウンカ日別誘殺数

## tabulaによる表の抽出

```python
dfs = tabula.read_pdf(
    "香川県農業試験場病害虫防除所_業務年報_2024.pdf",
    pages = '94',
    lattice=True
    )

# for df in dfs[0:-2]: # list内の1要素が1つの表に対応（dfs[0]と同義）
#     display(df)
# print(type(dfs[0]))
df = dfs[0]


df.drop([0, 1, 8, 9, 10], inplace=True)

df.columns = [
    '日',
    '4月_メス', '4月_オス',
    '5月_メス', '5月_オス',
    '6月_メス', '6月_オス',
    '7月_メス', '7月_オス',
    '8月_メス', '8月_オス',
    '9月_メス', '9月_オス',
    '10月_メス', '10月_オス',
    '11月_メス', '11月_オス'
    ]

df
```

```python
# すべての列を対象に分割
for col in df.columns:
    df[col] = df[col].str.replace(',', '').str.split('\r')

# 各行で「月日」の長さに合わせて他の列をパディング
for col in df.columns:
    max_len = df['日'].map(len) # 各行の「日」の長さ
    new_column = []
    for lst, ml in zip(df[col], max_len):
        if len(lst) < ml:
            new_lst = lst + [np.nan] * (ml - len(lst))
        else:
            new_lst = lst
        new_column.append(new_lst)
    df[col] = new_column

# 各セルがリストになったので，まとめて縦方向にexplode
df = df.explode(list(df.columns), ignore_index=True)

df
```

## 誤記チェック

```python
# print(df.dtypes)
# print('---')
df = df.astype('float64') # object -> float64
# print(df.dtypes)

df.sum()
```

## CSVとして保存

```python
df.to_csv("セジロウンカ日別誘殺数_2024.csv")
```
